```python
df2['date']=pd.to_datetime(df2['date'])
```

```python
df2.info()
```

### Extract Year

```python
df2['date_year']=df2['date'].dt.year
```

### Extract Month

```python
df2['date_month']=df2['date'].dt.month
```

### Extract days

```python
df2['date_days']=df2['date'].dt.day
```

### Extract Day of week

```python
df2['date_day_of_week']=df2['date'].dt.dayofweek
```

### Extract name of the day

```python
df2['date_name_of_day']=df2['date'].dt.day_name()
```

### Extract week number of the year

```python
df2['Week_number']=df2['date'].dt.isocalendar().week
```

### Extract quarter

```python
df2['quarter']=df2['date'].dt.quarter
```

### Extract Semester

```python
df2['Semester']=np.where(df2['quarter'].isin([1,2]),1,2)
```
